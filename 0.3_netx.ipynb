```python
# fast examination of all edges in a graph
FG = nx.Graph()
FG.add_weighted_edges_from([(1,2,0.125),(1,3,0.75),(2,4,1.2),(3,4,0.375)])
for n, nbrs in FG.adjacency_iter():
    for nbr, eattr in nbrs.items():
        data = eattr['weight']
        if data < 0.5: print('(%d,%d,%.3f)' % (n, nbr, data))
```

# Create Graphs
## see
http://networkx.lanl.gov/reference/generators.html

```python
# ER (number of nodes, p of connection, seed)
ER =nx.erdos_renyi_graph(10,0.5)
# WS (no of nodes, each n is con to k nearest, p of con, seed)
WS =nx.watts_strogatz_graph(10,2,0.5)
# BA (n of nodes, m No ed to attach from a new node to existing nodes, seed)
BA =nx.barabasi_albert_graph(30,2)
# RLB (exp n of nodes, p1 add edge to backbone, p2 add edge beyond back)
RLB =nx.random_lobster(10,0.9,0.5)
```
